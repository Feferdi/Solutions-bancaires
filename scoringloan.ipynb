#### Step 2: Create a function which plots 5 histograms

```python
# Create a function which plots 5 histograms - one for each variable.
def distributions(df):
    fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,1, figsize=(8,8))
    sns.distplot(df["age"], ax=ax1)
    sns.distplot(df["campaign"], ax=ax2)
    sns.distplot(df["cons_price_idx"], ax=ax3)
    sns.distplot(df["cons_conf_idx"], ax=ax4)
    sns.distplot(df["nr_employed"], ax=ax5)
    plt.tight_layout()
```

#### Step 5: Dimensionality reduction

```python
pca = PCA()
X_pca = pca.fit_transform(cluster_scale)
```
